## Corelation between Life Expectancy & Gross Domestic Product (GDP)


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv("Life_Expectancy_Data.csv")
```

```python
#converting mortality to Average_mortality and adding it to Avg_Life_Expertancy
data["Average_mortality"] = data["Adult Mortality"].mean()
data["Avg_Life_Expertancy_Mortality"] = data["Average_mortality"] + data.iloc[:,3]
data.info()
```

```python
#plot scatter plot to show corelation between life expectancy and population
x = data.iloc[:,23]
y = data.iloc[:,17]
fig, ax = plt.subplots()
ax.scatter(x, y)
plt.ylabel("Avg. Life Expectancy & Mortality")
plt.xlabel("Population")
plt.title("Life Expectancy & Population Correlation")
plt.show
```

```python
data1 = data.iloc[:,[23,17]]
correlation_matrix = data1.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Life Expectancy & Population Correlation")
plt.show
```

```python
#plot scatter plot to show corelation between life expectancy and GDP
x = data.iloc[:,23]
y = data.iloc[:,16]
fig, ax = plt.subplots()
ax.scatter(x, y)
plt.ylabel("Avg. Life Expectancy & Mortality")
plt.xlabel("GDP")
plt.title("Life Expectancy & GDP Correlation")
plt.show
```

```python
data1 = data.iloc[:,[23,16]]
correlation_matrix = data1.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Life Expectancy & GDP Correlation")
plt.show
```
