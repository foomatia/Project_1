# Step 2: Organising the data

```python
# Define the mapping of the countries per region
region_mapping = {
    "Africa": ["Algeria", "Angola", "Benin", "Botswana", "Burkina Faso", "Burundi", "Cabo Verde", "Cameroon", "Central African Republic", "Chad", "Comoros", "Djibouti", "Equatorial Guinea", "Eritrea", "Ethiopia", "Gabon", "Ghana", "Guinea", "Guinea-Bissau", "Kenya", "Lesotho", "Liberia", "Madagascar", "Malawi", "Mali", "Mauritania", "Mauritius", "Morocco", "Mozambique", "Namibia", "Niger", "Nigeria", "Rwanda", "Sao Tome and Principe", "Senegal", "Seychelles", "Sierra Leone", "South Africa", "Swaziland", "Togo", "Tunisia", "Uganda", "Zambia", "Zimbabwe"],
    "Europe": ["Albania", "Armenia", "Austria", "Azerbaijan", "Belarus", "Belgium", "Bosnia and Herzegovina", "Bulgaria", "Croatia", "Cyprus", "Estonia", "France", "Georgia", "Germany", "Greece", "Hungary", "Iceland", "Ireland", "Italy", "Latvia", "Lithuania", "Luxembourg", "Malta", "Montenegro", "Netherlands", "Norway", "Poland", "Portugal", "Romania", "Russian Federation", "Serbia", "Slovakia", "Slovenia", "Spain", "Sweden", "Ukraine"],
    "Oceania": ["Australia", "Fiji", "Kiribati", "Marshall Islands", "Micronesia", "Nauru", "New Zealand", "Palau", "Papua New Guinea", "Samoa", "Solomon Islands", "Tonga", "Tuvalu", "Vanuatu"],
    "Asia": ["Afghanistan", "Armenia", "Azerbaijan", "Bangladesh", "Bhutan", "Cambodia", "China", "Georgia", "India", "Indonesia", "Iraq", "Israel", "Japan", "Jordan", "Kazakhstan", "Kyrgyzstan", "Laos", "Lebanon", "Malaysia", "Maldives", "Mongolia", "Myanmar", "Nepal", "North Korea", "Oman", "Pakistan", "Palestine", "Philippines", "Qatar", "Saudi Arabia", "Singapore", "South Korea", "Sri Lanka", "Syria", "Syrian Arab Republic", "Tajikistan", "Thailand", "Timor-Leste", "Turkey", "Turkmenistan", "United Arab Emirates", "Uzbekistan", "Vietnam", "Yemen"],
    "Caribbean": ["Belize", "Dominican Republic", "Grenada", "Haiti", "Jamaica", "Saint Kitts and Nevis", "Saint Lucia", "Saint Vincent and the Grenadines", "Trinidad and Tobago"],
    "Central America": ["Costa Rica", "El Salvador", "Guatemala", "Honduras", "Mexico", "Nicaragua", "Panama"],
    "South America": ["Argentina", "Brazil", "Chile", "Colombia", "Ecuador", "Guyana", "Paraguay", "Peru", "Suriname", "Uruguay"],
    "North America": ["Canada", "United States"]
}

# Add a "Region" column to the DataFrame based on the region_mapping dictionary
datafile_df["Region"] = datafile_df["Country"].apply(lambda x: next((region for region, countries in region_mapping.items() if x in countries), "Unknown"))

# Display the DataFrame with the "Region" column
datafile_df.head()
```

```python
# Save the new data into CSV    
datafile_df.to_csv('datafile_per_region.csv', index=False)
```

```python
# Count of the number of countries per region
value_counts = datafile_df["Region"].value_counts()

print(value_counts)
```

```python
# Filter the DataFrame to include only data from 2004 to 2014
filtered_df = datafile_df[(datafile_df["Year"] >= 2004) & (datafile_df["Year"] <= 2014)]

filtered_df
```

```python
print(filtered_df.columns)
```

```python
# Clean column titles
new_column_names = {'infant deaths': 'Infant Deaths',
                    'percentage expenditure': 'Percentage Expenditure',
                    'Measles ': 'Measles',                 
                    ' BMI ': 'BMI', 
                    'under-five deaths ': 'Under-Five Deaths', 
                    'Total expenditure': 'Total Expenditure',
                    ' HIV/AIDS': 'HIV/AIDS',
                    'Income composition of resources': 'Income Composition of Resources'}

# Remove leading and trailing whitespaces from column names
filtered_df.columns = filtered_df.columns.str.strip()

# Create a copy of the DataFrame before renaming the columns
filtered_df_copy = filtered_df.copy()

# Rename the columns in the copied DataFrame
filtered_df_copy.rename(columns=new_column_names, inplace=True)

# Print the modified DataFrame
filtered_df_copy.head()
```

```python
# Verify that the columns are correct 
# and that the ' thinness  1-19 years' and ' thinness 5-9 years' columns have been dropped
print(filtered_df_copy.columns)
```

# Step 3: Creation of two separate DataFrames for analysing the countries per region more effectively

```python
# Creation of the DataFrame 1 including the 'Africa', 'Asia' and 'Europe' regions

grouped = filtered_df.groupby('Region')
region_dataframes = {region: group for region, group in grouped}

# List of regions to include
regions_to_include = ['Africa', 'Asia', 'Europe']

# Use pd.concat to concatenate DataFrames from the specified regions
dataframe1 = pd.concat([region_dataframes[region] for region in regions_to_include])

dataframe1.head()
```

```python
# Save the new DataFrame 1 in a CSV format
dataframe1.to_csv('datafile1_per_region.csv', index=False)
```

```python
# Creation of the DataFrame 2 including the 'North America', 'Central America', 'South America' and 'Caribbean' regions

# List of regions to include
regions_to_include = ['North America', 'Central America', 'South America', 'Caribbean']

# Use pd.concat to concatenate DataFrames from the specified regions
dataframe2 = pd.concat([region_dataframes[region] for region in regions_to_include])

dataframe2.head()
```

```python
# Save the new DataFrame 2 in a CSV format
dataframe2.to_csv('datafile2_per_region.csv', index=False)
```
