```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv("Life_Expectancy_2004-2014.csv")
data.head()
```

```python
#plot scatter plot to show corelation between life expectancy and population
x = data.iloc[:,3]
y = data.iloc[:,-5]
fig, ax = plt.subplots()
ax.scatter(x, y)
plt.xlabel("Life Expectancy")
plt.ylabel("Population")
plt.title("Life Expectancy & Population Correlation")
plt.show
```

```python
data1 = data.iloc[:,[3,17]]
correlation_matrix = data1.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Life Expectancy & Population Correlation")
plt.show
```

The above images show that there is a weak negative correlation between life expectancy and population in the dataset. This means that when life expectancy increases, population tends to decrease, but in a weak or unreliable manner.

```python
#plot scatter plot to show corelation between life expectancy and GDP
x = data.iloc[:,3]
y = data.iloc[:,-6]
fig, ax = plt.subplots()
ax.scatter(x, y)
plt.xlabel("Life Expectancy")
plt.ylabel("GDP")
plt.title("Life Expectancy & GDP Correlation")
plt.show
```

```python
data1 = data.iloc[:,[3,16]]
correlation_matrix = data1.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Life Expectancy & GDP Correlation")
plt.show
```

The above images show that there is a weak positive correlation between life expectancy and population in the dataset. This means that although both life expectancy and GDP tend to go up in response to one another, the relationship is not very strong.
